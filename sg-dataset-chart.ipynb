```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_excel("data.xlsx");

ax = sns.barplot(data=df, x='nama_data', y='value')

ax.set_facecolor("white");
ax.set_xlabel("Tahun")
ax.set_ylabel("MW")
ax.set_title("Data Kapasitas PLT ELB")
fig = ax.get_figure()
fig.set_facecolor("white")
plt.show()
```
